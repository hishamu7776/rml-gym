```python
import gym
from rmlgym.rmlgym import RMLGym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
import stable_baselines3.common.results_plotter as plotter
from stable_baselines3.common.callbacks import EvalCallback
```

```python
config_path = './examples/pendulum_keep_up.yaml'
```

```python
env = RMLGym(config_path)
```

```python
model = PPO("MlpPolicy", env, learning_rate=1e-4,batch_size=64, n_epochs=10, gamma=0.995, gae_lambda=0.97, clip_range=0.2,verbose=1 , tensorboard_log="./ppo_pendulum_rml/")
```

```python
history = model.learn(total_timesteps=1000000)
```

```python
model.save("ppo_pendulum_rml_model")
```

```python
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
```

```python
print(f"Mean Reward : {mean_reward}")
print(f"Std Reward : {std_reward}")
```
